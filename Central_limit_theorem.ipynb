```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# For reproducibility
np.random.seed(42)

# 1.  non-normal population
population = np.random.uniform(low=0, high=10, size=100000)
```

```python

# 2. Parameters
sample_sizes = [1, 5, 30, 100]
num_samples = 10000
```

```python
# 3. Plot
plt.figure(figsize=(10, 6))

for n in sample_sizes:
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.choice(population, size=n, replace=True)
        sample_means.append(np.mean(sample))

    plt.hist(
        sample_means,
        bins=50,
        density=True,
        alpha=0.5,
        label=f"Sample size n = {n}"
    )

# 4. Labels and title
plt.title("Central Limit Theorem Simulation", fontsize=14)
plt.xlabel("Sample Mean", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 400)
y = norm.pdf(x)

z = 1.96

plt.figure()
plt.plot(x, y)
plt.fill_between(x, y, where=(x <= z))
plt.fill_between(x, y, where=(x >= -z))
plt.axvline(z)
plt.axvline(-z)

plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution")
plt.show()
```
