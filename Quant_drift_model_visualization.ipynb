```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic returns (random walk with drift)
np.random.seed(42)
T = 500
mu = 0.0005      # drift
sigma = 0.01     # volatility

returns = np.random.normal(mu, sigma, T)
```

```python
# Rolling window size
N = 20
```

```python
# Compute Mt
Mt = []
for t in range(N, T):
    drift = np.sum(returns[t-N:t])
    Mt.append(drift / (sigma * np.sqrt(N)))

Mt = np.array(Mt)
```

```python

# Plot for visualization
plt.figure(figsize=(10, 4))
plt.plot(Mt, label="Normalized Drift $M_t$")
plt.axhline(1, linestyle='--', color='green', label="Bull Threshold")
plt.axhline(-1, linestyle='--', color='red', label="Bear Threshold")
plt.axhline(0, color='black')

plt.title("Risk-Scaled Drift Signal")
plt.xlabel("Time")
plt.ylabel("$M_t$")
plt.legend()
plt.grid(True)
plt.show()
```
